```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
# Load the data from the CSV file
file_path = 'C:\\Users\\AKASH SINGH YADAV\\Downloads\\householdtask3.csv'
df = pd.read_csv(file_path)

# Generate a bar chart for total households by year
plt.figure(figsize=(10, 6))
plt.bar(df['year'], df['tot_hhs'], color='green', label='Total Households')
plt.xlabel('Year')
plt.ylabel('Total Households')
plt.title('Total Households by Year')
plt.legend()
plt.show()
```

```python
# Generate a line chart for average income and expenditure by year
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['income'], marker='o', color='blue', label='Average Income')
plt.plot(df['year'], df['expenditure'], marker='o', color='red', label='Average Expenditure')
plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.title('Average Income and Expenditure by Year')
plt.legend()
plt.show()
```
